```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips') #loading inbuild dataset
tips
```

```python
#line plot:
sns.lineplot(x='total_bill',y='tip',data=tips)
plt.title('Relationship between tips and totalbill')
plt.show()
```

```python
#barplot
sns.barplot(x='day',y='total_bill',data=tips)  #here we dont need to groupby them.
plt.title('Day wise total bill')
plt.show()
```

```python
#scatter plot:-finding the correlation between tips and total_bill.
sns.scatterplot(x='total_bill',y='tip',data=tips,markers='*')
plt.title('Relationship between tips and totalbill')
plt.show()
```

```python
#boxplt:gererally used to findout the frequency of feature and outlier detection.
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
#violin plot:-
sns.violinplot(x="day",y='total_bill',data=tips)
```

```python
#heat map for correlation between features:-
sns.histplot(tips['tip'],bins=5,kde=True)
plt.title('Frequency of tips')
plt.show()
```

```python
sns.kdeplot(tips['tip'])
plt.title('Frequency of tips')
plt.show()
```

```python
#Heat map for correlation analysis:-
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='cool')
plt.show()
```
